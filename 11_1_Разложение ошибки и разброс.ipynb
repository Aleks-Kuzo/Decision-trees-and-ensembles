```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.datasets import fetch_california_housing
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
```

```python
RANDOM_STATE = 42
```

```python
df = fetch_california_housing(as_frame=True)

x = df.data
y = df.target
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.3, random_state=RANDOM_STATE)
```

```python
scaler = StandardScaler()
scaler.fit(xtrain)

xtrain = pd.DataFrame(data = scaler.transform(xtrain), columns = x.columns)
xtest = pd.DataFrame(data = scaler.transform(xtest), columns = x.columns)
```

С помощью bias variance decomposition оценим разложение ошибки

```python
from mlxtend.evaluate import bias_variance_decomp

x_train = xtrain.values
x_test = xtest.values
y_train = ytrain.values
y_test = ytest.values
```

```python
avg_mse, avg_bias, avg_var = bias_variance_decomp(LinearRegression(),
                                                    x_train, y_train, 
                                                    x_test, y_test, 
                                                    loss='mse', 
                                                    random_seed=np.random.seed(RANDOM_STATE))
```

```python
print(f'avg loss: {avg_mse}')
print(f'avg bias: {avg_bias}')
print(f'avg var: {avg_var}')
```

```python
avg_mse, avg_bias, avg_var = bias_variance_decomp(DecisionTreeRegressor(),
                                                    x_train, y_train, 
                                                    x_test, y_test, 
                                                    loss='mse', 
                                                    random_seed=np.random.seed(RANDOM_STATE))
```

```python
print(f'avg loss: {avg_mse}')
print(f'avg bias: {avg_bias}')
print(f'avg var: {avg_var}')
```

Смещение у решающего дерева в два раза меньше, но разброс намного больше. Значит, решающее дерево хорошо предсказывает переменную, но без подбора гиперпараметров оно сильно переобучено
